```python
from matplotlib import rcParams, cycler
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.ion()
```

```python
dataseturl="https://raw.githubusercontent.com/errjak/dataset/main/202004-divvy-tripdata.csv"
data=pd.read_csv(dataseturl)
```

```python
data.head(4)
```

```python
def diss (n,m):
  a = 2
  b = 0
  for col in ['start_lat','end_lat']:
    if data[col][n] == data[col][m]:
      b+=1
  return (a-b)/a
```

```python
print(diss(1,1))
print(diss(1,2))
print(diss(4,4))
```
