### Menghitung Dissimialirity

```python
def diss (n,m):
  a = 2
  b = 0
  for col in ['start_lat','end_lat']:
    if data[col][n] == data[col][m]:
      b+=1
  return (a-b)/a
```

```python
print(diss(1,1))
print(diss(1,2))
print(diss(4,4))
```

### DISKRITISASI

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
sepalLength = data.sepal_length
sepalWidth = data.sepal_width
petalLength = data.petal_length
petalWidth = data.petal_width
spesies = data.species
```

```python
aa = sepalLength.values.reshape(-1,1)
ab = sepalWidth.values.reshape(-1,1)
ac = petalLength.values.reshape(-1,1)
ad = petalWidth.values.reshape(-1,1)
sp = spesies.values.reshape(-1,1)
```

```python
est = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform')
```

```python
est.fit(aa)
est.fit(ab)
est.fit(ac)
est.fit(ad)
```

```python
xa = est.transform(aa)
xb = est.transform(ab)
xc = est.transform(ac)
xd = est.transform(ad)
```
